# What is the largest element of an array

```python
arr = [10, 20, 4, 45, 99]

largest = float('-inf')
second_largest = float('-inf')

for num in arr:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print(second_largest)
```

O(n) -> one pass through the array

# Second largest element in an array

```python
arr = [10, 20, 4, 45, 99]

largest = float('-inf')
second_largest = float('-inf')

for num in arr:
    if num > largest:
        second_largest = largest
        largest = num
    elif num > second_largest and num != largest:
        second_largest = num

print(second_largest)
```

Time Complexity:

# Check if the Array is Sorted

```python
arr = [1, 2, 3, 4, 5]

is_sorted = True
for i in range(len(arr) - 1):
    if arr[i] > arr[i + 1]:
        is_sorted = False
        break

print("Sorted" if is_sorted else "Not Sorted")
```

# Remove Duplicates from sorted array

```python
arr = [1, 1, 2, 2, 3]

if len(arr) == 0:
    print(0)

i = 0
for j in range(1, len(arr)):
    if arr[j] != arr[i]:
        i += 1
        arr[i] = arr[j]

# i + 1 is the length of unique elements
print("New length:", i + 1)
print("Updated array:", arr[:i+1])
```
