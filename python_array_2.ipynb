# Left rotate an array by one place

```python
# Brute
arr = [1,2,3,4,5,6]
temp_arr = arr[0]

left_rotate= arr[1:] + [temp_arr]

print(left_rotate)
```

Time complexity -> O(n)
Not in-place (uses extra space)

```python
# Optimal
arr = [1, 2, 3, 4, 5]

temp = arr[0]

for i in range(1, len(arr)):
    arr[i - 1] = arr[i]

arr[-1] = temp

print(arr)
```

Time complexity -> O(n)
Space complexity -> O(1)

# Left rotate the array by D spaces

```python
def reverse(arr, start, end):
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1
```

```python
arr = [1, 2, 3, 4, 5]
D = 2
n = len(arr)

D = D % n  # handle D > n

def reverse(arr, start, end):
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1

# Step 1
reverse(arr, 0, D - 1)

# Step 2
reverse(arr, D, n - 1)

# Step 3
reverse(arr, 0, n - 1)

print(arr)
```
